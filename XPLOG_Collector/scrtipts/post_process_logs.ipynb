```python
# prompt: read a file from drive line by line

FILE_NAME = "/content/drive/MyDrive/analysis_uk_work/logs_for_time_buffersize/1000Req_Multihost_XLP_For_Time.log"
reqID_wise_dict = {}
req_count=0
with open('{}'.format(FILE_NAME), 'r') as f:
  request_order_lines = []
  for line in f:
    if "Request Order" in line:
      try:
        req_count+=1
        request_order_lines.append(line.strip())
        # print(request_order_lines[0][9:])
        my_dict = json.loads(str(request_order_lines[req_count-1][9:-1]))
        host_pid = my_dict['event_context']['task_context']['host_pid']
        host_tid = my_dict['event_context']['task_context']['host_tid']
        ts = my_dict['event_context']['ts']

        my_id_string = my_dict['data']['lms'][-24:]
        # print("my_id_string = ",my_id_string)
        id_match = re.search(r'ID : (\d+)', my_id_string)
        id_value = id_match.group(1)

        my_id_string = my_dict['data']['lms']
        id_match = re.search(r'Request Order : (\d+)', my_id_string)
        Req_order = id_match.group(1)
        # print(Req_order)
        req_id = id_value
        if req_id not in reqID_wise_dict.keys():
          reqID_wise_dict[req_id] = {}
          reqID_wise_dict[req_id]["req_id"]= req_id
          reqID_wise_dict[req_id]["ts"]= ts
          reqID_wise_dict[req_id]["XPLOG_ID"]=(host_pid,host_tid)
          # print("host_pid :  {}, host_tid :  {}".format(host_pid,host_tid))
          reqID_wise_dict[req_id]["XPLOG_ID_{}".format(Req_order)]='''"host_pid" :  {}, "host_tid" :  {}'''.format(host_pid,host_tid)

          reqID_wise_dict[req_id]["sample_logs"]=[]
          reqID_wise_dict[req_id]["logs"]=[]

          reqID_wise_dict[req_id]['sample_logs'].append(line.strip())
        else:
          reqID_wise_dict[req_id]['sample_logs'].append(line.strip())
          # print("host_pid :  {}, host_tid :  {}".format(host_pid,host_tid))
          if req_count%6 ==0:
            ri = 6
          else:
            ri = req_count%6
          reqID_wise_dict[req_id]["XPLOG_ID_{}".format(Req_order)]='''"host_pid" :  {}, "host_tid" :  {}'''.format(host_pid,host_tid)
          # print("DONE")
      except:
        print("ERROR in = ",req_count, " == ",line.strip())
        # pass

      # break
req_count
```
